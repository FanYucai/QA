```python
filepath = r'C:\Users\Admin\Desktop\hmk\develop.data'
dataheader = ['Q', 'A', '?']
df = pd.read_csv(filepath, sep='\t', names=dataheader)
print(df.shape[0])
```

```python
df_cut = pd.DataFrame()
import jieba.posseg as pseg
for row in df.head().itertuples():
    strQ = " ".join(jieba.cut(row[1]))
    segments = pseg.cut(row[1], HMM=True)
    strQv = ""
    for w in segments:
        strQv = strQv + w.flag + ' '
        
    strA = " ".join(jieba.cut(row[2]))
    segments = pseg.cut(row[2], HMM=True)
    strAv = ""
    for w in segments:
        strAv = strAv + w.flag + ' '
    
    new = pd.DataFrame({"Q":strQ, "Qv": strQv, "A":strA,"Av": strAv, "?":row[3]}, index=["0"])
    df_cut = df_cut.append(new, ignore_index=True)

df_cut
```
